```python
import requests
import time
import matplotlib.pyplot as plt

api_url = "http://api.open-notify.org/iss-now.json"

def data_fetch():
    latitude_list, longitude_list, timestamp_list = [], [], []

    for _ in range(100):
        response = requests.get(api_url)
        data = response.json()

        latitude = float(data['iss_position']['latitude'])
        longitude = float(data['iss_position']['longitude'])
        timestamp = data['timestamp']

        print(longitude)
        print(latitude)
        print(timestamp)
        
        latitude_list.append(latitude)
        longitude_list.append(longitude)
        timestamp_list.append(timestamp)

        time.sleep(10)
        
    return latitude_list, longitude_list, timestamp_list


def graph_plotter(latitude_list, longitude_list):
    plt.figure(figsize=(10, 6))
    plt.plot(longitude_list, latitude_list, 'b.-')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('International Space Station Path')
    plt.grid(True)
    plt.show()

latitude_list, longitude_list, _ = data_fetch()
graph_plotter(latitude_list, longitude_list)
```
